# Map

Uma função é aplicada a cada elemento de uma lista resultando em uma nova lista de mesmo comprimento que a original.

```python
# Lista original
frutas = ["laranja", "abacaxi", "pêssego", "morango"]

# Função que será usada para a operação de 'map'
def para_maiusculas(palavra):
    return palavra.upper()
```

aplicando a funnção a cada elemento da lista usando um loop *for* (para cada)

```python
for palavra in frutas:
    fruta_maiuscula = para_maiusculas(palavra)
    print(fruta_maiuscula)
```

mas isso não nos retorna a lista!

```python
def frutas_em_maiusculas(lista_de_frutas):
    for palavra in frutas:
        fruta_maiuscula = para_maiusculas(palavra)
        print(fruta_maiuscula)
```

```python
frutas_em_maiusculas(frutas)
```

```python
# esta função vai fazer um 'map' propriamente
#porque ela vai criar uma NOVA lista mapeando
# os dados de lista de entrada
def frutas_maiusculas(lista_de_frutas):
    nova_lista = list()

    for palavra in frutas:
        fruta_maiuscula = para_maiusculas(palavra)
        nova_lista.append(fruta_maiuscula)

    return nova_lista
```

```python
frutas_maiusculas(frutas)
```

```python
for fruta in frutas_maiusculas(frutas):
    print(fruta)
```

# map()

```python
# Lista original
frutas = ["laranja", "abacaxi", "pêssego", "morango"]

# Função que será usada para a operação de 'map'
def para_maiusculas(palavra):
    return palavra.upper()
```

```python
map(para_maiusculas, frutas)
```

```python
list(map(para_maiusculas, frutas))
```

```python
for fruta_maiuscula in map(para_maiusculas, frutas):
    print(fruta_maiuscula)
```
